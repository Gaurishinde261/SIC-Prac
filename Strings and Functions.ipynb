```python
# Nested function
#one fun inanother is called decorater
```

```python
def greet():
    return "welcome"

def message(m):
    print(m.upper())

message(greet())
```

```python
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
@my_decorator
def sayhello():
    print("Hello")
sayhello()
```

```python
def mydefault(name = "Gana",age=21):
    print(f"My name is {name} and {age} year old")

mydefault()
mydefault("Falguni",25)
```

```python
def greet(age,name):
    print("Happy birthday")
    print(f"My age is {age} and my name is {name} ")
greet(name = "Ishita", age=22)
```

```python
#var argument
def total(*number):
    print(sum(number))

total(3,4,5,6)
total(34,98)
```
