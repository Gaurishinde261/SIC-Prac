## Quick Sort:
--quick sort is a sorting algorithm used to arrange elements (like numbers) in ascending or descending order.

--Work using Divide and conquer approach


--__Divide:__ split problem into smaller parts.

--__conquer:__ solve each smaller part.

--__combine:__ put the solution toghter to get the final answer.




```python
def quick_sort(numbers):
    if len(numbers) <= 1:
        return numbers
    #Pivot element
    pivot = numbers[len(numbers) // 2]
    left = [x for x in numbers if x < pivot]
    middle = [x for x in numbers if x == pivot]
    right = [x for x in numbers if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)
    


numbers = [56,43,23,90,12,21,55]
sorted_list = quick_sort(numbers)
print(sorted_list)


```

# Linear Search
--Linear search is sequential search algorithm where we start from one end and check every element of the list until the desired element is found .it is simplest searching algorithm

###### Linearsearch(array,key)
###### for each item in array
###### if item ==  value
###### return its index



```python
# Linear Search example 
def sequential_search(nlist,key):
    pos = 0
    found = False

    while pos < len(nlist) and not found:
        if nlist[pos] == key:
            found = True
        else:
            pos += 1 
    return found,pos
print(sequential_search([45,98,23,12,99,34,22], 10))
print(sequential_search(["Mango","Cherry","Strawberries","Grapes","Pineapple","Apple","Guava"], "Grapes"))
```

```python
def search_using_for(n,key):
    for index in range(len(n)):
        if n[index] == key:
            return index
    return False, -1

n = [76,34,23,12,90,76]
result = search_using_for(n,12)
print(result)
```

# Binary Search
--Binary search is a searching algorithm for finding an elements position in sorted array.

--In this approach,the element is always searched in the middle of portion of an array.

Binary search Working in two ways:

1.iterative Method

2.Recursive Method

--The recursive method follow the divide and conquer approch.

```python
def binary_using_itertive(array,target):
    low  = 0
    high = len(array) - 1
    while low <= high:
        mid = (low + high) // 2

        if array[mid] == target:
            return True,mid
        elif array[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return False, -1

array = [3,4,5,6,7,8,99]
print(binary_using_itertive(array,6))
print(binary_using_itertive(array,10))
```

```python
# Using Recursive
def binary_using_recursive(arr,target,low,high):
    if low > high:
        return False,-1

    mid = (low + high) // 2
    if arr[mid] == target:
        return True,mid
    elif arr[mid] < target:
        return binary_using_recursive(arr,target,mid + 1,high)
    else:
        return binary_using_recursive(arr,target,low,mid - 1)

arr = [4,5,6,7,8,9,10]
print(binary_using_recursive(arr,8,0,len(arr) - 1))
```

```python
#Making a dice game using pandas
import pandas as pd
import random

#set players
players=["Gopal","Govind"]

#set rounds
rounds=5

df=pd.DataFrame(index=players)

#roll the dice for each round
for r in range(1,rounds+1):
    df[f"Round {r}"]=[random.randint(1,6) for _ in players]

print(df)

#calculate total
df["Total"]=df.sum(axis=1)
print(df)

#Print Winnder
winner=df["Total"].idxmax()
print(f"The Winner is {winner}")

```
