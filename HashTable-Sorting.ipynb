```python
# Example of k mod n
# Search Key
keys =[24,52,91,67,48,83]

# size of hash table
n = 9

# Create empty hash Table
hash_table = [None] * n
print(hash_table)

# Hash Function
def hash_function(k):
    return k % n

# insert keys into hash table
for key in keys:
    index = hash_function(key)
    hash_table[index] = key
    print(f"key: {key} stored at index {index}")

print("-----Final Hash Table-----")
for i in range(n):
    print(f"Index {i}: {hash_table[i]}")
```

```python
# Hash Collsion Example

# Search key
keys = [24,52,91,67,48,83,67]

#Size of hash Table
n = 9

# create hash table
hash_table = [None] * n

#Hash Function
def hash_function(k):
    return k % n

#insert keys using Linear probing
for  key in keys:
    index = hash_function(key)
    original_index = index
    steps = 0

    # handle collsion
    while hash_table[index] is not None and steps < n:
        index = (index + 1) % n
        steps += 1
    if steps < n:
        hash_table[index] = key
        print(f"Key: {key} stored at index {index}")
    else:
        print(f"Hash table full! Could not insert key {key}")
        
print("-----Final Hash Table-----")
for i in range(n):
    print(f"Index {i}: {hash_table[i]}")
```

## 1.Bubble sort

--Bubble sort is sorting algo that compares two adjacent elements and swaps them until they are  in the intended order.

--Just like the movement of air bubble in the water that rise up to surface,each element of array move to the end i each iteration.therfore ,it is bubble sort

## 2.Selection Sort
--selection sort algo that selctsthe smallest element from an unsorted list in each iteration and place that element at begnning of unsorted list.

## 3. Insertion sort
-- Insertion sort is sorting algo that places an unsorted element at its sutabile place in each iteration.
-- For.ex:Insertion sort work similarly as we sort cards in our hand in card game.

## 4.Merge Sort
--Merge sort is one of the most poplular sorting algorithm that is based on the principal of __divide and conqure__ algorithm

-- here a problem is divided into multiple sub problems.Each sub-problem is slved individually.Finally,subproblems are combined to form the final solution



```python
# Bubble sort example
# count the swap


def bubble_sort(arr):
    n = len(arr)
    swap_count = 0

    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j],arr[j + 1] = arr[j + 1],arr[j]
                swap_count += 1
        return arr,swap_count
        

numbers = [45,32,12,89,45]
sorted_list,swaps = bubble_sort(numbers)
print("Sorted List:",sorted_list)
print("Numbers of swaps:",swaps)
```

```python
# Selection sort agorithm
def selectionSort(nlist):
    for fillslot in range(len(nlist) -1,0 -1):
        maxpos = 0
        for location in range(1,fillslot + 1):
            if nlist[location] > nlist[maxpos]:
                maxpos = location
        temp = nlist[fillslot]
        nlist[fillslot] = nlist[maxpos]
        nlist[maxpos] = temp
        
l =[14,96,43,27,41,21,99,70]
selectionSort(l)

print("Sorted List:",l)
```

```python
# Insertion sort
def insertionsort(l):
    for index in range(1,len(l)):
        currentvalue = l[index]
        position = index

        while position >0 and l[position - 1] > currentvalue:
            l[position] = l[position - 1]
            position = position - 1

        l[position] = currentvalue

l1 = [56,83,21,44,90,43]
print(l1)
insertionsort(l1)
print(l1)
print("\n")
l2 = ['A','B','c','a','T','o']
print(l2)
insertionsort(l2)
print(l2)
print("\n")
l3 = ['Ishita','Uma','Pallavi','Amruta','Aman']
print(l3)
insertionsort(l3)
print(l3)
```

```python

#write a function to perform partial selection sort on a floating point list and then verify the sorted order of the list
def insertion_sort_float(l):
    for index in range(1, len(l)):
        current_value = l[index]
        position = index

        while position > 0 and l[position - 1] > current_value:
            l[position] = l[position - 1]
            position -= 1

        l[position] = current_value

def verify_sorted_order(l):
    for i in range(len(l) - 1):
        if l[i] > l[i + 1]:
            return False
    return True

l = [12.5, 11.3, 13.7, 5.2, 6.8]
insertion_sort_float(l) 
print("Sorted list is:", l)
print("Is l sorted?", verify_sorted_order(l))

l2 = [3.14, 2.71, 1.41, 1.73, 0.577]
insertion_sort_float(l2)
print("Sorted list is:", l2)
print("Is l2 sorted?", verify_sorted_order(l2))
```

```python
# merge sort example
def mergesort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])
    return merge(left,right)

def merge(left,right):
    result = []
    i = j = 0
    while i< len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

    
arr = [56,43,12,90,32,76]
sorted_arr = mergesort(arr)
print(sorted_arr)
```
