# Greedy approach

- A Greedy algorithm biulds a solution step by step always choosing that looks best at moment(locally optimal) ,hoping this leads to globally optimal solution.

- Greedy Choice:
--At each step,pick the best immediate option without reconsidering past choices.
    - Fast and simple
    - Does not always guarantee optimal solution for all problemns.


```python
# make the change for an amount using fewest coins
def coin_change(coins,amount):
    coins.sort(reverse = True)
    result = []

    for coin in coins:
        while amount >= coin:
            amount -= coin
            result.append(coin)

    return result
print(coin_change([1,5,10,25], 63))
print(coin_change([1,5,10,20], 100))


# make the change with largest coin
def large_coin(amount):
    coins = [25, 10, 5, 1]

    count = 0
    for coin in coins:
        count += amount
        print(count)
        amount %= coin
        print(amount)
        
    return count 
    
print(large_coin(63))
```

# Dynamic programming
- Dynamic programming(DP)  is technique to solve problems by breaking them into overlapping subproblems and storing their results o each is solved only once.

- Dp is usually implemnted in follwing ways:

1. Top Down DP(REcusrion + memoization)
   - Easy to write
   - closely follow the mathematical defination
   - Risk of recursion depth issues for large n

2. Bottom up dp(tabulation)
   - no recusion
   - faster and safer for large inputs

3. 0/1 Knaps DP Patterns
   - each items can be chosen at most once(0 or 1 time)
   - you optimize something(max or min)
   - where is capicity/constraint(weight ,time ,cost etc)
     

```python
# Top -Down dp
def fibonacci_Top(n,memo={}):
    #base case
    if n <= 1:
        return n

    # if already compute,return from memo
    if n in memo:
        return memo[n]

    #compute and store in memo
    memo[n] = fibonacci_Top(n - 1,memo) + fibonacci_Top(n - 2,memo)
    #print(memo[n])
    print(memo)
    return memo[n]

print(fibonacci_Top(10))
```

```python
# Bottom -up dp
def fibonacci_Bottom(n):
    # Base Case
    if n <= 1:
        return n

    # Dp table
    dp =[0] * (n + 1)
    dp[0] = 0
    dp[1] = 1

    # build the table
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
        print(dp[i])
    return dp[n]

print(fibonacci_Bottom(10))
```

```python
# 0/1 knapsack

items =["A","B","C"]
weights = [1,3,4]
values = [15,20,30]

# knapack capactiyy
capacity = 4

# Dp array
dp = [0] * (capacity + 1)

# Create dp table
for i  in range(len(items)):
    for w in range (capacity,weights[i] -1,-1):
        dp[w] = max(dp[w], dp[w - weights[i]] + values[i])

print("Maximun value capacity", dp[capacity])
```

```python
# find maximum element using divide and conquer
def Find_max(arr,i,r):
    if i == r :
        return arr[i]

    mid = (i + r) // 2
    left_max = Find_max(arr,i,mid)
    right_max = Find_max(arr,mid+1,r)
    
    return left_max if left_max > right_max else right_max

l = [3,1,8,9,2]
print(Find_max(l,0,len(l) - 1))
```

```python
# sum of all elements using divide and conquer

def Find_sum(arr, i, r):
    if i == r:
        return arr[i]

    mid = (i + r) // 2
    left_sum = Find_sum(arr, i, mid)
    right_sum = Find_sum(arr, mid + 1, r)

    return left_sum + right_sum

l = [3, 1, 8, 9, 2]
print(Find_sum(l, 0, len(l) - 1))
```

# Backtracking

- Backtracking is prob solving techqnique

   - Buid a solution step by step
   - Abandon(backtrack) a path as soon as you know it cant lead to a valid solution
   - try the next possiblity

- Its commonly used for combinational problem like premutations,subsets,Sudoku,n-Queens



-- state: current partial solution
-- choice:what you try next
-- undo_choice:restore stse before next

```python
#Generate all permutations using backtracking algorithms
def permutation(nums):
    result = []

    def backtrack(path, used):
        if len(path) == len(nums):
            result.append(path[:])
            return

        for i in range(len(nums)):
            if used[i]:
                continue

            used[i] = True
            path.append(nums[i])

            backtrack(path, used)

            used[i] = False
            path.pop()

    backtrack([], [False] * len(nums))
    return result
# Example usage
nums = [1, 2, 3]
print(permutation(nums))
```
