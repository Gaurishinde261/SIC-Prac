```python
# 1. ACCESSING THE VALUE IN DICTIONARY 
my_dict={'name':'Alice','age':30,'city':'New York'}
print(my_dict['name'])          # Accessing value using key      
print(my_dict['age'])    # Accessing value using key       
```

```python
# 2. adding or updating key-value pair in dictionary
my_dict={'name':'Alice','age':30,'city':'New York'} 
my_dict['age']=31               # Updating existing key
my_dict['profession']='Engineer' # Adding new key-value pair   
print(my_dict)
```

```python
# 3. simply assigning the value to key in dictionary 
my_dict={'name':'Alice','age':30,'city':'New York'}
my_dict['age']=32               # Assigning new value to existing key   
print(my_dict)
```

```python
# remove item 
my_dict={'name':'Alice','age':32,'city':'New York'}
del my_dict['city']            # removing key-value pair using del keyword
print(my_dict)
```

```python
# using pop() method to remove item and get its value
my_dict={'name':'Alice','age':32}
removed_value = my_dict.pop('age')
print(removed_value)
```

```python
my_dict={'name':'Alice','age':32}
age = my_dict.pop('age')   # using pop() method to access value
print(my_dict) 
```

```python
# checking if a key exists in dictionary: you can check using 'in' keyword if a key exists 
my_dict={'name':'Alice','age':32}
if 'age' in my_dict:
    print("Key 'age' exists in the dictionary.")    
else:
    print("Key 'age' does not exist in the dictionary.")    
    
```

```python
my_dict={'name':'Alice','age':32, 'city':'pune'}
keys = my_dict.keys()   # getting all keys using keys() method  
values = my_dict.values() # getting all values using values() method    
print(keys)
print(values)   
```

```python
# clearing all items from dictionary: remove all items from the dictionary, use clear() method  
my_dict ={'name':'Gayatri','age':'32'}
my_dict.clear()
print(my_dict)
```

```python
#imp merging the dictionary: you use the update() method 
dict1 = {'name':'John'}
dict2 = {'age':23}
dict1.update(dict2)   # merging dict2 into dict1
print(dict1)        
```

```python
# nested dictionary: you have dictionaries inside other dictionaries 
nested_dict = {
    'person1': {'name': 'Alice', 'age': 30},
    'person2': {'name': 'Bob', 'age': 25}
}   
print(nested_dict)
c=nested_dict['person1']['name']  # accessing nested dictionary value
print(c)    
```

```python
#imp get function 
#Getting the default values: uses the get() method to access a value and avoid KeyError.
#if a key is not found(dosen't exist), it returns None or a specified default value instead of raising an error.
a={'name':'Gayatri','age':32}
my_dict = {'name':'Gayatri','age':'32','city':'pune'}
name = my_dict.get('name','unknow')  # accessing existing key
#if 'name' dosen't exist it will return 'unknown' as default value
na = my_dict.get('na')
print(na)
print(name)
```

```python
#copy dictionary 
original_dict = {'name':'Alice','age':30}
copied_dict = original_dict.copy()   # creating a shallow copy using copy() method  
print(copied_dict)
```
