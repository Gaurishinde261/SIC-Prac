# Grade

```python
score = int(input('enter score:'))
if score >= 90:
    grade='A'
    
elif score >= 80:
    grade ='B'
if score >= 70:
    grade='C'
if score >= 60:
    grade='D'
else:
    grade ='F'
print('your grade is:',grade)
```
