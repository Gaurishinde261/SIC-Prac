# lambda function 

## Defination:
#### lambda function is function defined using the lambda keyword instead of def and and it can contain only one expression 


# case study
**company analyze customer trancsction dta to
--remove invalid records 
--create new calculated fields
--prepare data for visulizations


```python
trans= [
{"customer":"A","amount":250,"status":"Success"},
{"customer":"B","amount":-50,"status":"Failed"},
{"customer":"C","amount":400,"status":"Success"},
{"customer":"D","amount":150,"status":"Success"}
]

# Filter invalid transactions
clean_data = list(filter(lambda x: x["status"] == "Success" and x["amount"] > 0 ,trans))
print(clean_data)

# transform data
# create a new colum:tax(10%)

transform_data = list(map(lambda x: {**x, "tax": x["amount"] * 0.9},clean_data))
print(transform_data)

"""def num(n):
    yield n
print(list(num(5)))"""

# Add Category "High" and "Low"
category = list(map(lambda x: {**x, "category": "High" if x["amount"] > 200 else "low"},transform_data)
        )

print(category)
#single line if statement
#result =50
#print("pass" if result >40 else "fail")

#Sort data
sorted_data = sorted(category,key=lambda x: x["amount"],reverse = True)
print(sorted_data)
```
