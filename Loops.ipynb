#  Random module 
# import module name1[, module name2,...]
Random Functions:
1.random():Genrate real Number between 0 to 1.
2.randrange():Return Int within specific range.
3.randint(a,b):Returns a random onteger N between a<=N<=b.
4.shuffle(seq):Randomly mixes element from given sequence list.
5.choice(seq):Selects any element in the sequence.
6.sample():Randomly Select a specified number of elements.

```python
import random as rd

print(rd.random())
print(rd.random())
print(rd.randrange(1,7))
print(rd.randrange(0,10,2))
print(rd.randint(1,10))
```
