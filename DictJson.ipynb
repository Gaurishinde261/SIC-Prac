```python
lst = [11,22,33,44,55]
print('berfore pop(0):',lst)
print('berfore pop(0) lst[1]=:',lst[1])
lst.pop(0)
print('after pop(0):',lst)
print('after pop(0) lst[1]=:',lst[1])
```

```python
dic = {0:11,1:22,2:33,3:44,4:55}
print('berfore pop(0):',dic.items())
print('berfore pop(0) dic[1]=:',dic[1])
dic.pop(0)
print('after pop(0):',dic.items())
print('after pop(0) dic[1]=:',dic[1])
```
