# Empolye salary Increment 
## increse salary by 10%

```python
#Method 1
salary={"A":30000,"B":40000}
for i in salary:
   salary[i]=salary[i]+(salary[i]*0.1)
print(salary)
```

```python
# Method 2
salary={"A":30000,"B":40000}
keys = list(salary.keys())
for i in keys:
    salary[i]+=0.1*salary[i]
print(salary)
```
