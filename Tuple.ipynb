```python
#access value in tuple using indexing 
my_tuple = (10, 20, 30, 40, 50)
print(my_tuple[2])  # Output: 30        
```

```python
#check existance of an item in tuple using if else 
my_tuple = (10, 20, 30, 40, 50)
if 30 in my_tuple:
    print("Item found")
else:
    print("Item not found")     
```

```python
#fetch the specific element using indexing 
my_tuple = (10, 20, 30, 40, 50)
print(my_tuple[2])  # Output: 30    
```

```python
#negative indexing in tuple
my_tuple = (10, 20, 30, 40, 50)     
print(my_tuple[-1])  # Output: 50   
```

```python
#slicing in tuple
my_tuple = (10, 20, 30, 40, 50, 60, 70)
print(my_tuple[2:5])  # Output: (30, 40,    
```
