# ***OOP Concepts***
## 1. Class:
-- A class is blueprint or templet for creating objects.

--It defines the properties(data) and methods(functions) an object will have.


## 2. Objects:
--An object is real instance of class.

--It represnts the properties and method defined by the class.

-- Objects use the properties and methods defined by the class.

### Q.Differnce between class and object
-- class is logical blueprint while an object is an physical instance from that blueprint

## 3. Method:
--A Method is a function that defined inside a class and is used to define the behaviour of an object.

### Types of Methods:

#### 1. Instance method

-- Works with object data
--use self

#### 2. Class method

-- Works with class-level data
-- Uses @classmethod
--Take cls as parameter

#### 3. Static method

---Does not use instancce or class data
--Uses @staticmethod

### Special(Magic) Methods
--__init__() -constructor

-- __str__() string representation

## 4.Encapsulation
-- restrict access to some of an objects internal details

1.Public member:accessible from anywhere
self.name = name

2.protected memeber(_variable): should not be accessed directly(convention only)
self._marks = marks

3.private members(__variable):name-managed to prevent direct access
self.__marks=marks

## 5.Inheritance
--inheritance concpet where one class (the child or derived class)acquire the properties of another class(th parent or base class).
using super() is used to call methods of parent class

### Types of inheritance
1.Single inheritance
2.Multiple inheritance
3.Multilevel inheritance
4.hierarchical inheritance
5.Hybrid inheritance

## 6.polymorphism
--polymorphism in python means "many forms" the same function name or operation behaves differently depen ding on object or context.

### Types:
1.Polymorphsim with functions
--The same function works with different datatypes

2. with operators
-- the same operator works differently for diff datatypess
   
3. with class methods(method ovrriding)
--Different classes have diff methods with same name but diff behavior

4.polymorphism using duck typing
-- python doesnt care about the object's types only whether it has the required method.

## 7.Composition(Has-A)
--composition(Has-A) where one class contains objects of another class,instead of inheriting from them.
--it models a"has-a" realtionship
EX.
inheritance(is-a):car is a vechicle
Composition(HAS-A):Car has an engine

--In composition,the contained object is usually created inside the owing class and used to provide functionality.

```python
# Example of Composition(HAS-A)

class Engine:
    def start(self):
        return "Engine started"


class wheel:
    def rotate(self):
        return "Wheel Rotating"

# Use composition in another class
class Car:
    def __init__(self):
        self.engine = Engine()
        self.wheel = [wheel() for _ in range(4)]

    def drive(self):
        print(self.engine.start())
        for wheel in self.wheel:
            print(wheel.rotate())
#Use Composition object
car = Car()
car.drive()
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass


# Concrete class
class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid Rs {amount} using Credit Card")


class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid Rs {amount} using PayPal")


# Using Abstraction
def make_payment(payment_method, amount):
    payment_method.pay(amount)


# Client code
credit = CreditCard()
paypal = PayPal()

make_payment(credit, 1000)
make_payment(paypal, 500)
```

```python
from abc import ABC, abstractmethod

class fuelError:
    pass

# composition
class engine:
    def __init__(self, horsepower, fuel):
        self.horsepower = horsepower
        self.fuel = fuel
    def start(self):
        return f'engine started with {self.horsepower} HP using {self.fuel}'

# abstraction
class auto(ABC):
    vehicle_count = 0
    def __init__(self, brand, model, price):
        self._brand = brand
        self._model = model
        self._price = price
        auto.vehicle_count += 1

    @abstractmethod
    def mileage(self):
        pass
    
    def get_details(self):
        return f'{self._brand}, {self._model} , ₹{self._price}'
    
    def vehicle_type(self):
        pass

    # using class method
    @classmethod
    def total_vehicle(cls): 
        return cls.vehicle_count

     #static method
    @staticmethod
    def company_policy():
        return "all cars runs"

class car(auto):
    def __init__(self, brand, model, price, engine, fuel):
        super().__init__(brand, model, price)
        self.engine = engine  # HAS-A Relationship
        self.fuel = fuel
        
    def mileage(self):
        return '15km/L'

    def vehicle_type(self):
        return "car"
    
    def refuel(self, liters):
        if liters <= 0:
            raise fuelError("invalid fuel amount")
        self.fuel += liters
        return f"refueled {liters} liters"

class ElectricCar(car):
    def mileage(self):      # polymorphism
        return "120km/kw"

    def vehicle_type(self):
        return "EV"

    def recharge(self, charge):
        if charge <= 0:
            raise fuelError("invalid charge amount")
        self.fuel += charge
        return f"recharged {charge} kW"

# operator overloading
class showroom:
    def __init__(self, name):
        self.name = name
        self.cars = []

    def addCar(self, other):
        self.cars.append(car)
    
    def __add__(self, other):
        newShowroom = showroom(self.name + " & " + other.name)
        return newShowroom
    
    def display(self):
        return [car.get_details() for car in self.cars]
    
    # polymorphuim function
    def printVehicleInfo(self, vehicle):
        print(vehicle.get_details())
        print("Type:", vehicle.vehicle_type())
        print("Mileage : ", vehicle.mileage())

# main program
if __name__ == "__main__":
    try:
        engine1 = engine(120, "Petrol")
        engine2 = engine(150, "Electric")

        car1 = car("Toyota", "Corolia", 150000, engine1, 40)
        car2 = ElectricCar("Tesla", "Model X", 600000,engine2, 0)

        engine1.start()
        engine2.start()


        showroom1 = showroom("City Motors")
        showroom2 = showroom("qwerty Motors")

        showroom1.printVehicleInfo(car1)
        showroom2.printVehicleInfo(car2)

        showroom1.addCar(car1)
        showroom2.addCar(car2)

        mega_showroom = showroom1 + showroom2

        for car in mega_showroom.display():
            print(car)

        print("Total Vehicle : ", auto.total_vehicle())
        print("Policy : ", auto.company_policy())

        print(car1.refuel(10))
        print(car2.recharge(20))

    except fuelError as e:
        print("Fuel Error: ", e)
```
