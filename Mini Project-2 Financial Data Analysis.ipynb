# Project Name: Financial Data Analysis (Stock dataset)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("symbols_valid_meta.csv")
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
df.info()
print("\nMissing Values:\n", df.isnull().sum())
```

```python
etf_counts = df['ETF'].value_counts()
print("ETF Counts:\n", etf_counts)

sns.barplot(x=etf_counts.index, y=etf_counts.values, palette='Set2')
plt.title("ETFs vs Non-ETFs")
plt.ylabel("Count")
plt.show()

```

```python
market_counts = df['Market Category'].value_counts()
print("Market Category Counts:\n", market_counts)

sns.barplot(x=market_counts.index, y=market_counts.values, palette='viridis')
plt.title("Distribution of Market Categories")
plt.xlabel("Market Category")
plt.ylabel("Number of Stocks")
plt.show()
```

```python
market_counts = df['Market Category'].value_counts()
print("Market Category Counts:\n", market_counts)

plt.pie(market_counts.values, labels=market_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Market Categories")
plt.show()
```

```python
# Companies with financial issue
problem_companies = df[df['Financial Status'] != '']
problem_companies[['Symbol','Security Name','Financial Status']]
```

```python
roundlot_counts = df['Round Lot Size'].value_counts()
print("Top Round Lot Sizes:\n", roundlot_counts.head(10))

sns.barplot(x=roundlot_counts.head(10).index, y=roundlot_counts.head(10).values, palette='coolwarm')
plt.title("Top 10 Round Lot Sizes")
plt.xlabel("Round Lot Size")
plt.ylabel("Number of Stocks")
plt.show()
```

```python
nasdaq_counts = df['Nasdaq Traded'].value_counts()
print("NASDAQ Traded vs Not Traded:\n", nasdaq_counts)

sns.barplot(x=nasdaq_counts.index, y=nasdaq_counts.values, palette='pastel')
plt.title("NASDAQ Traded vs Not Traded")
plt.ylabel("Count")
plt.show()
```

```python
top10_companies = df[['Symbol','Security Name']].sort_values('Symbol').head(10)
top10_companies
```

```python
etf_market = df[df['ETF']=='Y']['Market Category'].value_counts()
print("ETFs per Market Category:\n", etf_market)

sns.barplot(x=etf_market.index, y=etf_market.values, palette='magma')
plt.title("ETFs per Market Category")
plt.ylabel("Count")
plt.show()
```
